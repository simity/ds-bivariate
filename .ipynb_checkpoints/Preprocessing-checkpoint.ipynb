```python
import pandas as pd
```

```python
dataset=pd.read_csv("Placement.csv")
```

```python
dataset
```

```python
dataset['ssc_p'].dtype
```

```python
for columnName in dataset.columns:
    print(columnName)
```

```python
quan=[]
qual=[]
for columnName in dataset.columns:
    print (columnName)
    if (dataset[columnName].dtype=='O'):
        print(qual)
        qual.append(columnName)
    else:
        print(quan)
        quan.append(columnName)
```

```python
quan
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan, strategy="0")
imp.fit(dataset[quan])
df=imp.transform(dataset[quan])
```

```python
df=pd.DataFrame(df,columns=quan)
```

```python
df
```

```python
cate=dataset[qual]
```

```python
cate
```

```python
dfcate=[df,cate]
dataset=pd.concat(dfcate,axis=1)
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.to_csv("Preplacement.csv",index=0)
```
